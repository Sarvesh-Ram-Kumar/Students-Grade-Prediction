# Support Vector Regression (SVR)

## Importing the libraries

```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

## Importing the dataset

```python
dataset = pd.read_csv('Student_performance_data _.csv')
X = dataset.iloc[:, 1:-2].values
y = dataset.iloc[:, -2].values
```

```python
print(X)
```

```python
print(y)
```

## Splitting the dataset into the Training set and Test set

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.25, random_state = 0)
```

```python
print(X_train)
```

```python
print(y_train)
```

```python
print(X_test)
```

```python
print(y_test)
```

## Training the SVR model on the whole dataset

```python
from sklearn.svm import SVR
regressor = SVR(kernel = 'rbf')
regressor.fit(X_train, y_train)
```

## Predicting the Test set results

```python
y_pred = regressor.predict(X_test)
for i in range(0,len(y_test)):
    y_test[i]=round(y_test[i],2)
for i in range(0,len(y_pred)):
    y_pred[i]=round(y_pred[i],2)
A=np.concatenate((y_pred.reshape(len(y_pred),1), y_test.reshape(len(y_test),1)),1)
for i in range(0,len(A)):
    print(A[i])
```

## Error in Prediction

```python
total_error=0
for i in range(0,len(A)):
    error=abs(A[i][0]-A[i][1])
    total_error+=error
Final_error=total_error/len(A)
print(Final_error)
```
